```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 예제용 방향 그래프 생성
G = nx.DiGraph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

# 각 엣지에 대한 가중치 (또는 특성값) 설정
edge_weights = { (1, 2): 2, (1, 3): 1.5, (2, 3): 3, (3, 4): 2.5 }

# 그래프 그리기
pos = nx.spring_layout(G)  # 노드의 위치 결정
edges = G.edges()

# 엣지 가중치에 따라 컬러 맵 생성
weights = [edge_weights[edge] for edge in edges]
cmap = plt.cm.get_cmap('YlOrRd')  # 컬러 맵 설정 (Yellow-Orange-Red)

# 그래프 그리기
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=weights, edge_cmap=cmap, width=2.0)

# 컬러 바 추가
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=min(weights), vmax=max(weights)))
sm.set_array([])
cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_label('Edge Weight')

plt.title('Directed Graph with Colored Edges')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Use seed when creating the graph for reproducibility
G = nx.random_geometric_graph(200, 0.125, seed=896380)
  
# Position is stored as node attribute 'pos' for random_geometric_graph
pos = nx.get_node_attributes(G, "pos")

# Find node closest to (0.5, 0.5) and its distance
dmin = 1.0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter, dmin = n, d

# Compute shortest path lengths from ncenter to all other nodes
p = nx.single_source_shortest_path_length(G, ncenter)

# Draw the graph with nodes colored by shortest path lengths
plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(G, pos, alpha=0.4)

# Nodes colored by shortest path lengths from ncenter
nx.draw_networkx_nodes(G, pos, nodelist=list(p.keys()), 
                       node_size=80, 
                       node_color=list(p.values()), 
                       cmap=plt.cm.Reds)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis('off')
plt.show()

# 어렵지 않은 코드인데 직접구현하라면 못하겠다. 애초에 nx문법에 익숙하지도 않고
#cmap인가 이런거로 채색등 뭔가 코딩 자체보다 라이브러리 지식의 문제가 큰듯
```

```python
import networkx as nx
import matplotlib.pyplot as plt

seed = 668237  # 무작위성 재현을 위한 시드 설정

z = [3, 3, 3, 3, 3, 2, 2, 2, 1, 1, 1]

# degree sequence가 그래프를 형성할 수 있는지 확인
print(nx.is_graphical(z))

print("Configuration model")
G = nx.configuration_model(z, seed=seed)  # configuration model 생성
degree_sequence = [d for n, d in G.degree()]
print(f"degree sequence: {degree_sequence}")

print("Degree histogram")
hist = {}
for d in degree_sequence:
    if d in hist:
        hist[d] += 1
    else:
        hist[d] = 1

print("Degree - #nodes")
for d in hist:
    print(f"{d:4} {hist[d]:6}")


# 그래프 그리기
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=seed)  # 무작위 위치 배치
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray', width=1.5)

plt.title("Graph with Degree Sequence")
plt.show()


# 노드의 위치 정보를 spring layout으로 설정
#pos = nx.spring_layout(G, seed=seed)

# 그래프 그리기 옵션 설정
#options = {
#    'font size': 12,
#    'node_size': 36,
#    'node_color': 'white',
#    'edge_color': 'black',
#    'linewidths': 2,
#    'width': 3,
#    'font_size': 12,
#}

#nx.draw_networkx(G, pos=pos, **options)
#nx.draw_networkx(G, pos=pos)


# 축과 레이블 설정
#ax = plt.gca()
 
#plt.title("Graph with Customized Options")
#plt.show()
```

```python
import networkx as nx

# 룰리팝 그래프 생성
G = nx.lollipop_graph(4, 6)

# 노드별 최단 경로 길이 계산
pathlengths = []
print("Source vertex -> {Target:length}")
for v in G.nodes():
    spl = dict(nx.single_source_shortest_path_length(G, v))
    print(f"{v} -> {spl}")
    for p in spl:
        pathlengths.append(spl[p])

# 평균 최단 경로 길이 계산
average_shortest_path_length = sum(pathlengths) / len(pathlengths)
print(f"Average shortest path length: {average_shortest_path_length:.2f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Karate club 그래프 생성
G = nx.karate_club_graph()

# Degree 정보를 DataFrame으로 변환
df = pd.DataFrame(list(G.degree), columns=['node', 'degree']).set_index('node')

# 각 노드의 클럽 소속 정보 추가
df['club'] = pd.Series({node: data['club'] for node, data in G.nodes(data=True)})

# 각 클럽별 최대 degree를 기준으로 색상 설정
df['color'] = df.groupby('club')['degree'].transform(lambda x: x / x.max())
df.loc[df['club'] == 'Officer', 'color'] *= -1

# 그래프 레이아웃 설정
layout = nx.fruchterman_reingold_layout(G, seed=2001)

# 그래프 그리기 설정
plt.figure(figsize=(12, 8))
options = {
    'font_size': 12,
    'node_size': 500,
    'linewidths': 2,
    'width': 1,
}

# 노드 색상 및 레이블 적용하여 그리기
nx.draw(G, pos=layout, node_color=df['color'], cmap=plt.cm.coolwarm, **options)

# 컬러바 추가
sm = plt.cm.ScalarMappable(cmap=plt.cm.coolwarm, norm=plt.Normalize(vmin=df['color'].min(), vmax=df['color'].max()))
sm.set_array([])
plt.colorbar(sm)

plt.title("Karate Club Graph")
plt.show()
```
