### 시그마 대수의 정의

주어진 집합 $\Omega$ 위의 Collection(시그마 대수) $F\$는 다음 세 가지 조건을 만족합니다:

1. $ \emptyset \in F \$

2. 만약 $ A \in F \$ 이면 \$ \Omega \setminus A \in F \$

3. 만약 \$ \{A_i\}_{i=1}^{\infty} \subseteq F \$ 이면 \$ \bigcup_{i=1}^{\infty} A_i \in F \$

의미 
시그마 대수는 셀 수 있는 것 잴 수 있는 것들의 집합을 수식으로 표현한것이다.
0과 전체는 셀 수 있다. 어떤 것을 셀 수 있다면 전체에서 뺀 것도 셀 수 있다.
셀 수 있는것에 셀 수 있는 것을 더하면 그것 또한 셀 수 있다.

### 확률 변수의 정의

확률변수는 표본공간  ($ \Omega $, $ \mathcal{F}$, $P$ )에서, <br>
$ \Omega $에서 $ \mathbb{R} $ 또는 $ \mathbb{R}^n $으로의 함수로 정의됩니다. 즉, $ X: \Omega \rightarrow \mathbb{R} $ 또는 $ X: \Omega \rightarrow \mathbb{R}^n $입니다.

여기서,

- \$ \Omega \$는 표본공간$sample space$으로, 가능한 모든 결과$표본$의 집합입니다.
- \$ \mathcal{F} \$는 \$ \Omega \$의 부분집합들을 포함하는 시그마 대수$sigma algebra$로, 이는 \$ \Omega \$의 모든 부분집합들에 대해 확률을 할당할 수 있는 구조를 제공합니다.
- \$ P \$는 확률 측도$probability measure$로, \$ \mathcal{F} \$에 정의된 사건들에 대해 확률을 할당하는 함수입니다.
- \$ X \$는 확률변수로, 각 표본 \$ \omega \in \Omega \$에 대해 실수값 $X$$(\omega$) 를 할당합니다.

이러한 정의에 따라, 확률변수는 표본공간의 각 표본에 대해 어떤 실수값을 할당하는 함수이며, 이는 확률 이론과 통계학에서 중요한 개념입니다.

============================================================================================================

### 확률변수의 성질

 
주어진 확률공간 ($\Omega, \mathcal{F}, P $ )에서 정의된 확률변수 $X$는 $\Omega$에서 실수 $\mathbb{R}$로의 함수로, 즉 $ X: \Omega \rightarrow \mathbb{R} $입니다. 이 함수는 다음과 같은 특성을 만족합니다:


- **각 표본에 대한 값 할당**: 각 표본 \$\omega \in \Omega\$에 대해 $X$($\omega$)는 실수 값입니다.


- **사건에 대한 확률 계산**: 임의의 실수 $x \in \mathbb{R}$에 대해,  $X$가 $x$보다 작거나 같은 값이 될 확률은 $P$($X$ $\leq x$)로 정의됩니다. 즉,

  $P$($X\leq x$) = P({$\omega \in \Omega$ : $X$($\omega$) $\leq$ $x$] )


- **확률분포 정의**: $X$의 확률분포는 $X$가 취할 수 있는 각 값 $x$에 대한 확률 $P$($X$ = $x$)를 나타냅니다. 이는 사건 { $\{X = x\}$의 확률로 정의됩니다

============================================================================================================

### 예시를 통한 공부
주사위 던지기 예시를 통한 확률변수와 확률 측도의 개념 설명:

표본공간 (\$\Omega\):
- 주사위를 던질 때 나올 수 있는 모든 가능한 결과들의 집합입니다. 주사위를 던지면 나올 수 있는 결과는 1부터 6까지의 숫자입니다. 따라서,
 $[
  \Omega = \{1, 2, 3, 4, 5, 6\}
  ]$

<br>

시그마 대수 $\mathcal{F}$:
- 표본공간 $\Omega$의 모든 부분집합들을 포함하는 시그마 대수입니다. 주사위를 던질 때 가능한 사건(부분집합)들은 예를 들어 다음과 같습니다:
  - 홀수가 나오는 사건: \$\{1, 3, 5\}\$
  - 짝수가 나오는 사건: \$\{2, 4, 6\}\$
  - 특정 숫자가 나오는 사건: \$\{3\}, \{5\}\$ 등

<br>


확률 측도 $P$:
- 주어진 시그마 대수 \$\mathcal{F}\$에 대해 정의된 확률 측도입니다. 주사위를 던져서 각 숫자가 나올 확률은 모두 같고, 각 숫자는 1/6의 확률을 가집니다. 따라서,

  \[
  P$\{1\}$ = P$\{2\}$ = P$\{3\}$ = P$\{4\}$ = P$\{5\}$ = P$\{6\}$ = $\frac{1}{6}
  $

확률변수 $X$:
- 주사위를 던져서 나오는 결과를 나타내는 함수입니다. 주사위의 결과를 실수값으로 매핑하는 확률변수 \$X\$를 정의할 수 있습니다. 예를 들어, 주사위를 던져서 나온 숫자를 그대로 실수로 표현하는 경우,


  $X$($\omega$) = $X$ $\quad \text{for }$ $\omega \in \Omega $
  
  여기서 $X$($\omega$)는 주사위를 던져서 나온 결과 $\omega$에 대응하는 실수값입니다. 예를 들어, 주사위를 던져서 3이 나왔을 때, 확률변수 $ X $는 $X(3)$ = 3 이 됩니다.


이제 확률변수 $ X $를 통해 다양한 확률 이론적인 계산을 할 수 있습니다. 예를 들어,

- 확률 $ P(X = 3)$ 은 주사위를 던져서 3이 나올 확률로, $P({3})$ = $\frac{1}{6} $입니다.
- 기대값 \$ \mathbb{E}[X] \$은 주사위를 던져서 나온 숫자들의 평균으로, \$ \mathbb{E}[X] = 3.5 \$입니다.
- 분산 \$ \mathrm{Var}$X$ \$은 주사위를 던져서 나온 숫자들의 분포에 대한 측도로, \$ \mathrm{Var}$X$ = \frac{35}{12} \$입니다.

이처럼 확률변수는 특정 사건에 대한 확률뿐만 아니라, 기대값, 분산 등 다양한 확률적 특성을 분석하는 데 유용하게 사용됩니다.

\$ E[X] = \sum_{x} x \cdot P(X = x) \$

여기서,

\$ \mathbb{E}[X] = \int_{-\infty}^{\infty} x f_X(x) \, dx \$


- \( x \)는 확률 변수 \( X \)가 취할 수 있는 모든 가능한 값들을 나타냅니다.
- \( P(X = x) \)는 확률 변수 \( X \)가 \( x \)라는 값을 가질 확률을 나타냅니다.

### Wasserstein 거리의 정의

1. 첫 번째 Wasserstein 거리 (\( W_1(P, Q) \)):
   $$ W_1(P, Q) = \inf_{\gamma \in \Gamma(P, Q)} \int_{\mathbb{R}^d \times \mathbb{R}^d} \| x - y \| \, d\gamma(x, y) $$

2. 두 번째 Wasserstein 거리 (\( W_2(P, Q) \)):
   $$ W_2(P, Q) = \left( \inf_{\gamma \in \Gamma(P, Q)} \int_{\mathbb{R}^d \times \mathbb{R}^d} \| x - y \|^2 \, d\gamma(x, y) \right)^{1/2} $$

# Kolmogorov-Smirnov 거리는 

두 확률 분포 함수 \( F(x) \)와 \( G(x) \) 사이의 최대 차이를 나타내는 지표

$[ D_{KS}(F, G) = \sup_x |F(x) - G(x)| ]$

여기서,
- \( F(x) \)는 첫 번째 확률 분포의 누적 분포 함수(CDF),
- \( G(x) \)는 두 번째 확률 분포의 누적 분포 함수(CDF)입니다.
- $\( \sup_x \)는 모든 \( x \) 값에 대해 최대값을 구하는 것을 의미합니다.$

이 수식은 두 CDF 사이의 절대적인 최대 차이를 계산하여 두 확률 분포의 유사성을 측정합니다. \( D_{KS} \)가 클수록 두 분포는 서로 다르다고 판단됩니다.
