```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Figure 생성
fig = plt.figure(figsize=(8, 4))

# Axes 생성
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # 왼쪽, 아래, 너비, 높이
ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])  # 왼쪽, 아래, 너비, 높이

# 각각의 Axes에 데이터 플롯
ax1.plot(x, y, 'b')
ax2.plot(x, np.cos(x), 'r')

# 축 (Axis) 레이블 설정
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')

# 축 (Axis) 제목 설정
ax1.set_title('Main Plot')
ax2.set_title('Inset Plot')

# 그래프 보여주기
plt.show()
#plt.plot(x, y, 'o') # plot circle markers
```

```python
import matplotlib.pyplot as plt

def make_patch_spines_invisible(ax):
    ax.set_frame_on(True)  # 축에 프레임을 켜는 것 무슨 의미인지 모르는데 우선 이정도로 이해
    ax.patch.set_visible(False)  # 배경을 조정하는 것 같다. False라서 숨기는거고
    for sp in ax.spines.values():  #  spines라는 건 아마 왼쪽 오른쪽  위쪽 아래쪽 값
        sp.set_visible(False)  # 여기 각각에 실시한다.
        
# figure 생성 및 축 생성
fig, host = plt.subplots()  # figure 객체와 주요 축을 생성한다는데 대충 도화지 만들고 중심축 잡는건가??
fig.subplots_adjust(right=0.75)  # 오른쪽 여백 조정 직관적으로 이해가네

par1 = host.twinx()  # host 축과 동일한 x축을 공유하는 첫 번째 보조 축을 생성합니다
par2 = host.twinx()  # host 축과 동일한 x축을 공유하는 두 번째 보조 축을 생성합니다

par2.spines["right"].set_position(("axes", 1.2))  # par2의 오른쪽 스파인 위치를 설정합니다
make_patch_spines_invisible(par2)  # par2의 배경을 숨기고, 스파인을 보이지 않게 설정합니다
par2.spines["right"].set_visible(True)  # par2의 오른쪽 스파인을 다시 보이도록 설정합니다
#어렵다. 보면서 대충은 알겠는데 내가 이정도를 직접 코딩하려면..




#그림 및 축 생성
fig, host = plt.subplots()  # figure 객체와 주요 축을 생성합니다
fig.subplots_adjust(right=0.75)  # 그림의 오른쪽 여백을 조정합니다

par1 = host.twinx()  # host 축과 동일한 x축을 공유하는 첫 번째 보조 축을 생성합니다
par2 = host.twinx()  # host 축과 동일한 x축을 공유하는 두 번째 보조 축을 생성합니다

par2.spines["right"].set_position(("axes", 1.2))  # par2의 오른쪽 스파인 위치를 설정합니다
make_patch_spines_invisible(par2)  # par2의 배경을 숨기고, 스파인을 보이지 않게 설정합니다
par2.spines["right"].set_visible(True)  # par2의 오른쪽 스파인을 다시 보이도록 설정합니다


#데이터 생성 및 입력
p1, = host.plot([0, 1, 2], [0, 1, 2], "b-", label="Density") #host.plot위에 한다 랑 색깔 라벨 만 제외하면 딱히 어려운건 없다.
p2, = par1.plot([0, 1, 2], [0, 3, 2], "r-", label="Temperature") #x축 데이터는 동일하게 통일 y축 데이터만 각자
p3, = par2.plot([0, 1, 2], [50, 30, 15], "g-", label="Velocity")

#축 설정하기
host.set_xlim(0, 2) #x축은 공유하기 때문에 host에 하나만 만들어주 면된다 나머지는 각각 공유
host.set_ylim(0, 2)
par1.set_ylim(0, 4)
par2.set_ylim(1, 65)

#축 레이블과 눈늠 설정하기
host.set_xlabel("Distance") #이것 또 한 마찬가지
host.set_ylabel("Density") 
par1.set_ylabel("Temperature")
par2.set_ylabel("Velocity")

#축 레이블 색상 및 눈금 설정하기
host.yaxis.label.set_color(p1.get_color()) #축 레이블 색상과 객체를 가져오는 것 같다는데 솔직히 잘 이해는 안 가네
par1.yaxis.label.set_color(p2.get_color())
par2.yaxis.label.set_color(p3.get_color())

tkw = dict(size=4, width=1.5) # 딕셔러니 타입을 이용해서 눈금 설정하기 그래프를 그릴때 딕셔러니로 하고 한번에 자주 넣는다. 눈금 스타일
host.tick_params(axis='y', colors=p1.get_color(), **tkw)# 눈금 지정하기
par1.tick_params(axis='y', colors=p2.get_color(), **tkw)
par2.tick_params(axis='y', colors=p3.get_color(), **tkw)
host.tick_params(axis='x', **tkw)

#범례추가
lines = [p1, p2, p3]  #왜 만드는가 싶었는데 아래에 for문으로 넣으려고
host.legend(lines, [l.get_label() for l in lines]) #이해가 될랑 말랄하네

#출력하기
plt.show()

#그냥 대충 이런거라고 알겠다.
#강의 특성상 이렇게 이해하는게 최선인것 같다. 애초에 이거만 제대로 이해하려고 해도 5~10분 넘게 설명해야하는데 강의에서는 1분컷
#1분이 아니라 한 30초 정도 했나? 그냥 딱 이런게 있다 싶은 입문용 강의가 맞고 제대로는 나중에 공부해야함
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#1. 라이브러리 임포트 및 데이터 준비 (무난하게 다 이해가는 부분)
N = 5  # 그룹의 수
menMeans = (20, 35, 30, 35, 27)  # 남성 그룹의 평균 값
womenMeans = (25, 32, 34, 20, 25)  # 여성 그룹의 평균 값
menStd = (2, 3, 4, 1, 2)  # 남성 그룹의 표준 편차
womenStd = (3, 5, 2, 3, 3)  # 여성 그룹의 표준 편차
#라이브러리 임포터 데이터 준비
ind = np.arange(N)  # x 축 위치를 설정하기 위한 배열
width = 0.35  # 막대의 너비

#2. stacked-bar 그래프 생성 (yerr) 오차막재 제외하고는 무난하게 이해
p1 = plt.bar(ind, menMeans, width, yerr=menStd)  # 남성 데이터 막대
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd)  # 여성
 
# 3. 눈금 등 기타등등  (눈금 설정함수 ticks 가 새롭네)
plt.ylabel('Scores')  # y 축 레이블 설정
plt.title('Scores by group and gender')  # 그래프 제목 설정
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))  # x 축 눈금 레이블 설정
plt.yticks(np.arange(0, 81, 10))  # y 축 눈금 설정
plt.legend((p1[0], p2[0]), ('Men', 'Women'))  # 범례 설정 P1,P2 의 데이터 양식은 line2d인데 이걸 아마 리스트처럼 쓸 수 있나보지?

plt.show()
#ticks랑 p1[0] 이런식으로 뭔가 출력하는 형식 그리고 yerr이 부분을 제외하면 특별하게 어려운점은 없었따.
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 1.사전 데이터 준비
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(labels))  # 레이블의 위치
width = 0.35  # 막대의 너비

#2. 그림 및 축 생성
fig, ax = plt.subplots()  # figure 객체와 축을 생성합니다 (아직도 이 부분이 자세히 이해가지 않는다.)
rects1 = ax.bar(x - width/2, men_means, width, label='Men')  # 남성 데이터 막대
rects2 = ax.bar(x + width/2, women_means, width, label='Women')  # 여성 데이터 막대

#3 그래프 설정
ax.set_title('Scores by group and gender')  # 그래프 제목 설정
ax.set_ylabel('Scores')  # y 축 레이블 설정
ax.set_xticks(x)
ax.set_xticklabels(labels)  # x 축 눈금과 레이블 설정
ax.legend()  # 범례 설정

# 데이터값 표시및 함수 정의
def autolabel(rects):
    for rect in rects: #성별 데이터 막대에서 데이터를 하나씩 받아온다. 뭐가 들어갔는지 신경을 안 썼네 역시 직접하는거랑 각주랑 다르네
        height = rect.get_height() # 음 막대 객체로 섦정한거고 거기에 height 라는 값이 자동으로 들어가보네 그러고보니까 난 객체를 생성
        ax.annotate('{}'.format(height), #annotate 좌표에 텍스트를 추가
                    xy=(rect.get_x() + rect.get_width() / 2, height), # 텍스트를 표시할 위치를 설명 get함수를 이용해서 이것저것 가져오네
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points", #좌표시스템 
                    ha='center', va='bottom') #가로 세로 역시

autolabel(rects1)  # rects1 (남성 데이터)에 대해 데이터 값을 표시합니다
autolabel(rects2)  # rects2 (여성 데이터)에 대해 데이터 값을 표시합니다

fig.tight_layout()  # 그림의 주변 공백을 최적화합니다 이게 이런 의미였군
plt.show()  # 그림을 화면에 출력합니다

#mapplotlib 어느정도 익숙해졌다고 생각했는데 이거 공부하며서 자신감 박살나는중 ..
```

```python
#코드 분석하면서 느낀건데 이거 제대로 설명하려면 1~3시간 정도 걸림
#이걸 15분만에 하려니까 이해가 안 가는게 당연 통계는 내가 배경지식이 있으니까 어느정도 이해 가능했고
# 그래프 파트 부분 이해 안 갔던것도 이런 맥락.... 이건 말 그대로 부스트코스처럼 입문용 강의인듯

#데이터 생성
x = np.linspace(0, 2 * np.pi, 400)  # 0부터 2*pi까지의 값을 400등분하여 생성
y = np.sin(x ** 2)  # x의 제곱에 대한 사인 함수 값을 계산

#서브플롯 생성 및 데이터 그리기
fig, axs = plt.subplots(2, 2)  # 2x2 서브플롯을 생성합니다 (이 부분이 이해가 안 갔던건 fig, axs가 무엇을 할당받는지 몰라서)

axs[0, 0].plot(x, y)  # 좌상단 서브플롯에 데이터를 그래프로 그립니다
axs[0, 0].set_title('Axis [0,0]')  # 좌상단 서브플롯의 제목을 설정합니다

axs[0, 1].plot(x, y, 'tab:orange')  # 우상단 서브플롯에 주황색으로 데이터를 그래프로 그립니다
axs[0, 1].set_title('Axis [0,1]')  # 우상단 서브플롯의 제목을 설정합니다

axs[1, 0].plot(x, -y, 'tab:green')  # 좌하단 서브플롯에 초록색으로 데이터를 그래프로 그립니다
axs[1, 0].set_title('Axis [1,0]')  # 좌하단 서브플롯의 제목을 설정합니다

axs[1, 1].plot(x, -y**2, 'tab:red')  # 우하단 서브플롯에 빨간색으로 데이터를 그래프로 그립니다
axs[1, 1].set_title('Axis [1,1]')  # 우하단 서브플롯의 제목을 설정합니다

#그래프 및 레이블 설정
fig.suptitle('Sharing x per column, y per row')  # 전체 그림의 제목을 설정합니다

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')  # 각 서브플롯의 x축과 y축 레이블을 설정합니다

for ax in axs.flat:
    ax.label_outer()  # 각 서브플롯의 레이블을 바깥쪽에만 표시하도록 설정합니다
    
#데이터 출력
plt.show()  # 그림을 화면에 출력합니다

#매일 무의식적으로 하던 plt.subplots()을 하면 어떤 변수가 할당되는지 피상적으로나마 알게해줌
```

```python
#원형 그래프 그리기
#figure와 ax
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
#앞장이랑 다르게 subplot에서 매소도를 본격적으로 활용
#특히 추가적인 키워드 인자 쓰는 부분이 새롭다. fig가 전체라면 AXes는 말 그대로 칸을 나누는 느낌

#2.데이터 준비 및 함수 설정 (리스트에서 for 문 써서 그거로 리스트 만드는 테크닉  이것만 제외하면 어렵지 않음)
recipe = ["375 g flour", "75 g sugar", "250 g butter", "300 g berries"]
data = [float(x.split()[0]) for x in recipe]  # 각 재료의 그램 수를 추출하여 리스트로 저장
ingredients = [x.split()[-1] for x in recipe]  # 각 재료의 단위를 추출하여 리스트로 저장

def func(pct, allvals): #내가 각주 달 때 함수에 받는 인자가 무슨 데이터양식인지 적을 필요는 없나? 너무 초보적인거라?
    absolute = int(pct / 100.0 * np.sum(allvals))  # 각 부분의 절대값을 계산
    return "{:.1f}%\n({:d} g)".format(pct, absolute)  # 퍼센트와 절대값을 포맷하여 반환

#3원형 그래프 그리기 (이건 autopoc  말그대로 백분율자동, 남자를 이용해서 장의한것 , 칼라로 딕셔러닝을이용해서 대충은 알겠는ㄷ)
wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),  # autopct로 func 함수를 사용하여 텍스트 설정
                                  textprops=dict(color="w"))  # 텍스트 속성 설정

#범례추가및 그래프(대충은 알기 쉽지만 정확히 알기는 어렵네)
ax.legend(wedges, ingredients, title="Ingredients",
          loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))  # 범례 설정

plt.setp(autotexts, size=8, weight="bold")  # 자동 생성된 텍스트의 크기와 굵기 설정
ax.set_title("Matplotlib bakery: A pie")  # 그래프 제목 설정

plt.show()  # 그림을 화면에 출력합니다
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse #처음보는데
import matplotlib.transforms as transforms #대충 뭔지는 알겠네

#이 함수에 당황해서 시간을 너무 많이 날렸다. 당황했을 때 침착성을 유지하는게 중요하다. 
def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs): #x,y=좌표, ax=xes , 편차, 색상, 추가적인 키워드 인자
    if x.size != y.size: 
        raise ValueError("x and y must be the same size") #오류 메시지 처리
    
    #공분산 공식 그냥 그대로 구현하기
    cov = np.cov(x, y)  # x와 y의 공분산 계산 
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])  # Pearson 상관 계수 계산
    
    #이 부분은 이해가 안 가는데 뭐 신뢰타원? 이런게 있다고만 이해하자
    ell_radius_x = np.sqrt(1 + pearson)  # x 방향 타원의 반지름
    ell_radius_y = np.sqrt(1 - pearson)  # y 방향 타원의 반지름
    
    
    #이것도 자세히는 모르지만 타원을 구현하는거라고 이해하자 당황하지말고 침착히
    ellipse = Ellipse((0, 0), width=ell_radius_x * 2, height=ell_radius_y * 2,
                      facecolor=facecolor, **kwargs)
    
    #이것도 분산을 이용해서 타원을
    scale_x = np.sqrt(cov[0, 0]) * n_std  # x의 표준 편차에 n_std를 곱하여 크기 조정
    scale_y = np.sqrt(cov[1, 1]) * n_std  # y의 표준 편차에 n_std를 곱하여 크기 조정
    
    #간단
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    #아핀변환 구현한거라고 생각
    transf = transforms.Affine2D().rotate_deg(45).scale(scale_x, scale_y).translate(mean_x, mean_y)
    
    #객이미 생성된 위에 아핀변환에서 트랜스
    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)


def get_correlated_dataset(n, dependency, mu, scale): # 생성할 변수의 수, 의존성을 나타내는 2x2행렬, 평균, 스케일
    latent = np.random.randn(n, 2)  # n개의 2차원 랜덤 샘플 생성
    dependent = latent.dot(dependency)  # 주어진 dependency 행렬과의 내적 계산
    scaled = dependent * scale
    scaled_with_offset = scaled + mu  # mu를 오프셋으로 사용하여 데이터 스케일링
    return scaled_with_offset[:, 0], scaled_with_offset[:, 1] #스케일 반환
#연산에 대해서 잘 모르니까 제대로 이해할 수 없다. 그냥 이런 연산을 통해서 저런값을 반환한다고 생각




#파라미터랑 기타등등 설정
np.random.seed(0)
PARAMETERS = {
        'Positive correlation': np.array([[0.85, 0.35],[0.15, -0.65]]),
        'Negative correlation': np.array([[0.9, -0.4],[0.1, -0.6]]),
        'Weak correlation': np.array([[1, 0],[0, 1]])
    }#
mu = 2, 4  # mean of x = 2; mean of y = 4
scale = 3, 5  # standard deviation of x = 3; standard deviation of y = 5

#맨날 하는 그거
fig, axs = plt.subplots(1, 3, figsize=(9, 3))


for ax, (title, dependency) in zip(axs, PARAMETERS.items()):  #itemms랑 zip대충은 아는건데 자세히는 모르는것들이네
    x, y = get_correlated_dataset(800, dependency, mu, scale) # 이거에 따라서 의존성 행렬을 만들고
    ax.scatter(x, y, s=0.5)  # scatter plot of random samples (x, y)
    ax.axvline(c='grey', lw=1) #축을 나타나낸다.
    ax.axhline(c='grey', lw=1)  # draw x & y axes
    confidence_ellipse(x, y, ax, edgecolor='red')  # draw confidence ellipse
    ax.scatter(mu[0], mu[1], c='red', s=3)  # plot center point (mu[0], mu[1])
    ax.set_title(title)  # set subplot title

plt.show()

    #어렵지만 이게 대충 무엇인지 정도는 이해할 수 있었다.
    #당황하지말자 침착하게.....
    #근데 진짜 시간 오래걸리네 이걸 15분만에 끝내니까.. 아니 아직도 분량이 남았지
```

```python
# 필요한 라이브러리 임포트
from mpl_toolkits.mplot3d import Axes3D #3차원 그래프 생성용
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator #눈금 생성용
import numpy as np

# Figure와 3D Axes 생성
fig = plt.figure()
ax = fig.gca(projection='3d') #3차원 코딩 위해서 gca라는 매소드 이용 이렇게 이해하자  아 현재 활성화된 AXes를 반환한다는게 이제 뭔지

# 데이터 생성
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) #모든조합 가능하게 하는거로 3차원 할때 이용하는거 이런 느낌
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# 색상 설정 (체스판 무늬 패턴)
colortuple = ('y', 'b')
colors = np.empty(X.shape, dtype=str)
xlen = len(X)
ylen = len(Y)
for y in range(ylen):
    for x in range(xlen):
        colors[x, y] = colortuple[(x + y) % len(colortuple)]

# 표면 플로팅
surf = ax.plot_surface(X, Y, Z, facecolors=colors, linewidth=0)

# 축 설정
ax.set_zlim(-1, 1)  # Z 축의 범위 설정
ax.w_zaxis.set_major_locator(LinearLocator(6))  # Z 축의 눈금 설정

# 그래프 표시
plt.show()
#딥러닝 연습할 때 몇번 구현해서 알겠는데 그때보다는 잘 알겠는 여전히 헷갈리는게 있네
```

```python
from string import ascii_letters #문자열 처리에 뭐하는거인듯 아스키코드를 바꿔주는 대충 그런 느낌?
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)), columns=list(ascii_letters[26:]))
#아스키문자열을 여기서 쓰네 데이터 열에 이름을 지정하기 위해서

# Compute the correlation matrix
corr = d.corr() #상관행렬

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool) #그냥 간단하게 상관해렬 크기의 0행렬
mask[np.triu_indices_from(mask)] = True #특정조건 만족하는 상삼각행렬로

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True) #색상 변수도 만드는게 가능한거였군

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()

#진짜 돌리고 코드 읽는데만 15분이 넘을듯 진짜 이렇게 보니까 새삼 그냥 이런 종류맀다고만 하고 넘어갔네

#보면 대충 이해가지만 내가 이걸 코딩하기는 ...
```
